#### Import some basic Libraries and Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

#### Importing the Training and Test set

```python
train=pd.read_csv("data/train.csv")
test=pd.read_csv("data/test.csv")
```

```python
print("Shape:",train.shape)
train.head()
```

```python
print("Shape:",test.shape)
train.head()
```

#### Combine both training and test file for data preprocessing and drop the SalePrice and ID columns

```python
data=train.drop(["SalePrice"],axis=1).append(test,sort=False)
data.drop("Id",inplace=True,axis=1)
data.info()
```

#### Handelling Missing Values 

```python
#Removing all the columns which have null values more than 40 percent
s=pd.Series(train.isnull().sum()>584)
most_null=s.index[s.values==True]
data.drop(most_null,inplace=True,axis=1)
data.info()
```

```python
#Finding the columns which have null values
s1=pd.Series(data.isnull().sum()>0)
null_columns=s1.index[s1.values==True]
data[null_columns].info()
```

#### Seperating the Categorical and Numerical columns

```python
cat_var=data.select_dtypes(include='object').columns
cat_info=data[cat_var].describe()
cat_info
```

```python
num_var=['MSSubClass', 'LotFrontage', 'LotArea','MasVnrArea','BsmtFinSF1',
             'BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
             'GrLivArea','GarageArea','WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch',
            'ScreenPorch', 'PoolArea', 'MiscVal']

num_info=data[num_var].describe()
num_info
```

```python
null_num_cat=['BsmtFullBath','BsmtHalfBath','GarageCars','GarageYrBlt']
```

#### Removing Outliers

```python
# Removing outliers for numerical values
for var in num_var:
    mean=num_info.loc['mean',var]
    std=num_info.loc['std',var]
    uppr_limit=mean+3*std
    lower_limit=mean-3*std
    data.loc[data[var]>=uppr_limit,[var]]=uppr_limit
    data.loc[data[var]<=lower_limit,[var]]=lower_limit
```

```python
# Filling the missing values
for var in cat_var:
    data[var].fillna(data[var].mode()[0],inplace=True)

for var in null_num_cat:
    data[var].fillna(data[var].mode()[0],inplace=True)

for var in num_var:
    data[var].fillna(data[var].mean(),inplace=True)
```

```python
data.isnull().sum().sum()
```

#### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
for var in cat_var:
    data[var] = labelencoder.fit_transform(data[var])
```

```python
yr_columns=['YearBuilt','YearRemodAdd','GarageYrBlt','YrSold']
for var in yr_columns:
    data[var] = labelencoder.fit_transform(data[var])
```

```python
data.head()
```

```python
features=data.columns
features
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data[num_var]=sc.fit_transform(data[num_var])
data.head()
```
